```python
import random
def enemies_alive(enemies):
  print(f'Enemy horde alive: ')
  for enemy in enemies:
    print(f"Name: {enemy['name']} ------- HP: {enemy['hp']}")

def fireball_damage():
  return sum(random.randint(1,6) for _ in range(8))

enemies = [
    {'name': 'Aboleth', 'dexterity': 3, 'hp': 100, 'resistance': '', 'immunities': ''},
    {'name': 'Orc', 'dexterity': 1, 'hp': 13, 'resistance': '', 'immunities': ''},
    {'name': 'Ghoul', 'dexterity': 2, 'hp': 22, 'resistance': '', 'immunities': 'Poison'},
    {'name': 'Fire Snake', 'dexterity': 2, 'hp': 22, 'resistance': '', 'immunities': 'Fire'},
    {'name': 'Bearded Devil', 'dexterity': 2, 'hp': 52, 'resistance': 'Fire', 'immunities': ''}

]

spells = [
    {'name': 'Fireball', 'cd': 17, 'damage': fireball_damage, 'type': 'Fire'}
]

enemies_alive(enemies)

used_spell = next((spell for spell in spells if spell['name']=='Fireball'), None)
total_damage = used_spell['damage']()
print(f'\nThe spell {used_spell['name']} has been used on the horde! Causing {total_damage} of fire damage!')

for enemy in enemies:
  if enemy['immunities'] == used_spell['type']:
    final_damage = 0
  else:
    final_damage = total_damage

  enemy_dex_save = random.randint(1,20) + enemy['dexterity']

  if enemy_dex_save >= used_spell['cd']:
    final_damage = final_damage // 2

  if enemy['resistance'] == used_spell['type']:
    final_damage = final_damage // 2

  enemy['hp'] -= final_damage

for enemy in enemies[:]:
    if enemy['hp'] <= 0:
        print(f"{enemy['name']} is dead!")
        enemies.remove(enemy)

print('\n')
enemies_alive(enemies)
```
