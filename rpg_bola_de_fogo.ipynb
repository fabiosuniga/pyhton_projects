```python
import random

enemy = [
    {'id' : 1, 'nome' : 'goblin jovem', 'hp' : 7},
    {'id' : 2, 'nome' : 'esqueleto', 'hp' : 20},
    {'id' : 3, 'nome' : 'cão infernal', 'hp' : 33},
    {'id' : 4, 'nome' : 'assassino', 'hp' : 60},
    {'id' : 5, 'nome' : 'civil', 'hp' : 3},
]

print(f"A Horda possui {len(enemy)} Inimigos\n\nInimigos restantes:")

for enemies in enemy:
    print(f"{enemies['nome']}: HP restante: {enemies['hp']}")

bola_de_fogo = sum(random.randint(1,6) for _ in range(8))

print(f"\nA bola de fogo deu {bola_de_fogo} de dano!")

for enemies in enemy[:]:
    enemies['hp'] -= bola_de_fogo
    if enemies['hp'] <= 0:
        enemy.remove(enemies)

print(f"\n\nA Horda possui {len(enemy)} Inimigos\n\nInimigos restantes:")

for enemies in enemy:
    print(f"{enemies['nome']}: HP restante: {enemies['hp']}")
```
