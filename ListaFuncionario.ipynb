```python
lista_funcionario = []

while True:
  nome = input(f"Digite o nome do funcionário: ")
  idade : int = int(input(f"Digite a idade do {nome}: "))
  profissao = input(f"Digite a profissão do {nome}: ")

  funcionario = {'nome' : nome, 'idade' : idade, 'profissao' : profissao}
  lista_funcionario.append(funcionario)

  opcao = input("Quer adicionar mais algum funcionário? S/N: ").upper()
  if opcao == 'N':
    break
  else:
    print("\nDados do próximo funcionário")

print("\nLista total dos funcionários")
for funcionario in lista_funcionario:
  print(f"Nome do funcionário: {funcionario['nome']}\nIdade: {funcionario['idade']}\nProfissão: {funcionario['profissao']}\n")




```
