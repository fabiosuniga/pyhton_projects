```python
milhas = [10, 50, 100, 2]

quilometros = list(map(lambda x: x*1.6 , milhas))
print(quilometros)
```

```python
placar = [
    {'nome': 'Player1', 'score': 1500},
    {'nome': 'NoobMaster', 'score': 500},
    {'nome': 'ProGamer', 'score': 3000}
]

placarOrdenado = sorted(placar, key= lambda x: x['score'], reverse = True)
print(placarOrdenado)
```

```python
goldPiece = int(input("How much golden pieces do you have?: "))

shopItems = {'Life Potion': 35, 'Sword': 100, 'Shield': 40}

for name, value in shopItems.items():
  print(f'Name: {name}, Value(GP): {value}')

if goldPiece >= shopItems['Life Potion']:
  goldPiece -= shopItems['Life Potion']
  print(f'\nLife potion aquired! your new gold balance are {goldPiece} gold pieces.')

else:
  print('Insuficient gold pieces for this puchase')
```
