CRIA UMA LISTA NOVA APENAS COM VALORES UNICOS

```python
def unique_list(lst):
  new_lst = []
  for x in lst:
    if x not in new_lst:
      new_lst.append(x)
  return new_lst

lst = [1,1,1,1,2,2,3,3,3,3,4,5]
print(unique_list(lst))
```

O CODIGO ABAIXO DIZ QUANTAS LETRAS MINUSCULAS E MAIUSCULAS O TEXTO POSSUI

```python
def up_low(s):
  low = 0
  high = 0

  for char in s:
    if char.isupper():
      high += 1
    elif char.islower():
      low += 1

  print(f'Numbers of low letters: {low} and High letter: {high}')

s = 'Hello Mr. Rogers, how are you this fine Tuesday?'
up_low(s)
```

O CODIGO ABAIXO CHECA SE O PRIMEIRO NUMERO DA FUNÇÃO ESTA NO ALCANCE DO MENOR E DO MAIOR PASSADO POR ELA TAMBÉM

```python
def ran_check(num, low, high):
  range_list = []
  for i in range(low, high+1):
    range_list.append(i)
  return num in range_list

ran_check(5,2,7)
```

O CODIGO RETORNA VERDADEIRO SE O NUMERO 0 0 7 APARECER NESTA ORDEM NA LISTA, NÃO NECESSARIAMENTE EM SEGUIDA

```python
number = [1,2,0,4,0,7,5]

def spy_game(numbers):
  code = [0,0,7,'x']
  for x in numbers:
    if x == code[0]:
      code.pop(0)

  return len(code) == 1

print(spy_game(number))
```

O CODIGO ABAIXO RETORNA APENAS RESULTADOS FORA DO ALCANCE COMEÇANDO EM 6 ATÉ 9, O RESTO ELE SOMA

```python
num = [4,5,6,7,8,9]

def summer_69(numbers):
  result = 0
  add = True

  for x in numbers:
    while add:
      if x != 6:
        result += x
        break
      else:
        add = False
      while not add:
        if x != 9:
          break
        else:
          add = True
  return result

print(summer_69(num))
```

O CODIGO ABAIXO RETORNA VERDADEIRO SE TIVER DOIS NUMEROS 3 EM SEQUENCIA

```python
numbers = [1,3,1,1,1,3,3]

def has33(list_num):
  for x in range(0, len(list_num)):
    if list_num[x] == 3 and list_num[x+1] == 3:
      return True
  return False

print(has33(numbers))

```

O CODIGO ABAIXO REVERTE A ORDEM DAS PALAVRAS DE UMA FRASE

```python
text = 'i am home'

def yoda_reverse(text):
  new_text = text.split()
  new_text = new_text[::-1]
  return ' '.join(new_text).capitalize()

print(yoda_reverse(text))
```

O CODIGO ABAIXO PEGA O FUNCIONARIO COM MAIS HORAS REGISTRADAS

```python
work_hours = [('Abby', 200), ('Fabio', 800), ('Pietra', 400)]

def best_employee(employee):
  max_hours = 0
  employee_top = ''

  for name,hours in employee:
    if hours > max_hours:
      max_hours = hours
      employee_top = name
  return (employee_top, max_hours)

print(best_employee(work_hours))
```

O CODIGO ABAIXO FILTRA APENAS NUMEROS PARES

```python
def even_list(num_list):
  new_list = []
  for number in num_list:
    if number % 2 == 0:
      new_list.append(number)
  return new_list

print(even_list([1,2,3,4,5,6]))
```

O CODIGO ABAIXO FILTRA O JOGADOR DO MELHOR PRO PIOR EM SCORE

O CODIGO ABAIXO CALCULA UMA VENDA DE ITEM

```python
goldPiece = int(input("How much golden pieces do you have?: "))

shopItems = {'Life Potion': 35, 'Sword': 100, 'Shield': 40}

for name, value in shopItems.items():
  print(f'Name: {name}, Value(GP): {value}')

if goldPiece >= shopItems['Life Potion']:
  goldPiece -= shopItems['Life Potion']
  print(f'\nLife potion aquired! your new gold balance are {goldPiece} gold pieces.')

else:
  print('Insuficient gold pieces for this puchase')
```
